```python
import numpy as np
import pandas as pd
from scipy.stats import median_abs_deviation

def calculateNNMD(logFC, positive_control, negative_control):
    positive_control = list(set(logFC.index).intersection(positive_control))
    negative_control = list(set(logFC.index).intersection(negative_control))
    
    logFC_positive_control = logFC.loc[positive_control]
    logFC_negative_control = logFC.loc[negative_control]
    
    median_diff = logFC_positive_control.median(axis=0) - logFC_negative_control.median(axis=0)
    mad_negative_control = median_abs_deviation(logFC_negative_control, axis=0, nan_policy='omit')
    nnmd = median_diff / mad_negative_control
    return nnmd,mad_negative_control

def calculateNNMD_zscore(logFC, positive_control, negative_control):
    positive_control = list(set(logFC.index).intersection(positive_control))
    negative_control = list(set(logFC.index).intersection(negative_control))
    
    logFC_positive_control = logFC.loc[positive_control]
    logFC_negative_control = logFC.loc[negative_control]
    
    mean_diff = logFC_positive_control.mean(axis=0) - logFC_negative_control.mean(axis=0)
    std_negative_control = logFC_negative_control.std(axis=0, ddof=0)
    
    nnmd = mean_diff / std_negative_control
    return nnmd

```

```python
NNMD,mad_negative_control = calculateNNMD(LFCSequencesPerTarget, RasCombinatoryCommonEssentialControls['GuideTargetSymbol'], RasCombinatoryNonEssentialControls['GuideTargetSymbol'])
NNMD_zscore = calculateNNMD_zscore(LFCSequencesPerTarget, RasCombinatoryCommonEssentialControls['GuideTargetSymbol'], RasCombinatoryNonEssentialControls['GuideTargetSymbol'])

# Calculate average representations
AvgRepresentations = RASFullCounts.mean()[1:] #remove pdna

# Create DataFrame with the calculated metrics
RASSequenceQualityMetrics = pd.DataFrame({
    'SequenceName': LFCSequencesPerTarget.columns,
    'NNMD': NNMD,
    'NNMD_zscore': NNMD_zscore,
    'AvgRepresentations': AvgRepresentations.values
})

print(RASSequenceQualityMetrics)
```

```python

from scipy.stats import spearmanr

cor_list = []
cur_screenmap=RASScreenMap

for i in range(len(cur_screenmap)):
    cur_model = cur_screenmap.iloc[i]['CellLine']
    cur_sample = cur_screenmap.iloc[i]['Screen']
    if cur_sample =='pDNA' or cur_model not in AvanaLogfoldChange6.columns:
        continue
    ov_genes = LFCScreensPerTarget.index.intersection(AvanaLogfoldChange6.index)

    avana_combined = pd.DataFrame({
        'target': ov_genes,
        'lfcRAS': LFCScreensPerTarget.loc[ov_genes, cur_sample],
        'lfcAvana': AvanaLogfoldChange6.loc[ov_genes, cur_model]
    })

    avana_combined['Essentiality'] = np.where(
        avana_combined['target'].isin(RasCombinatoryCommonEssentialControls['GuideTargetSymbol']),
        'Pan-essential genes',
        np.where(
            avana_combined['target'].isin(RasCombinatoryNonEssentialControls['GuideTargetSymbol']),
            'Non-essential genes',
            'Others'
        )
    )
    avana_combined=avana_combined[~(avana_combined['lfcAvana'].isna() | avana_combined['lfcRAS'].isna())]
    
    cor_coef, _ = spearmanr(avana_combined['lfcRAS'], avana_combined['lfcAvana'])
    cor_list.append(pd.DataFrame({'Screen': [cur_sample], 'CellLine': [cur_model], 'SpearmanAVANA': [cor_coef]}))

# Combine the correlation results into a single DataFrame
cor_df = pd.concat(cor_list, ignore_index=True)

# Join the correlation results with the RASScreenMap DataFrame
RASScreenMap = RASScreenMap.merge(cor_df, on=['Screen', 'CellLine'], how='left')

# Print the resulting DataFrame
print(RASScreenMap)
```

```python

from matplotlib.ticker import MaxNLocator
os.makedirs('QualityControl/UsePlatepDNA/CompareToAVANA',exist_ok=True)


for i in range(len(cur_screenmap)):
    cur_model = cur_screenmap.iloc[i]['CellLine']
    cur_sample = cur_screenmap.iloc[i]['Screen']
    if cur_sample =='pDNA' or cur_model not in AvanaLogfoldChange6.columns:
        continue
    ov_genes = LFCScreensPerTarget.index.intersection(AvanaLogfoldChange6.index)

    avana_combined = pd.DataFrame({
        'target': ov_genes,
        'lfcRAS': LFCScreensPerTarget.loc[ov_genes, cur_sample],
        'lfcAvana': AvanaLogfoldChange6.loc[ov_genes, cur_model]
    })

    avana_combined['Essentiality'] = np.where(
        avana_combined['target'].isin(RasCombinatoryCommonEssentialControls['GuideTargetSymbol']),
        'Pan-essential genes',
        np.where(
            avana_combined['target'].isin(RasCombinatoryNonEssentialControls['GuideTargetSymbol']),
            'Non-essential genes',
            'Others'
        )
    )

    
    min_lim = np.floor(avana_combined[['lfcRAS', 'lfcAvana']].min().min())
    max_lim = np.ceil(avana_combined[['lfcRAS', 'lfcAvana']].max().max())
    lims = (min_lim, max_lim)
    
    plt.figure(figsize=(7, 7))
    
    
    plt.fill_betweenx([lims[0], -1], lims[0], -1, color='#fbf7f7',edgecolor='grey')
    plt.fill_between([-1, lims[0]], lims[0], -1, color='#fbf7f7',edgecolor='grey')
    
    plt.axhline(0, linestyle='--', color='k')
    plt.axvline(0, linestyle='--', color='k')
    plt.plot(lims, lims, 'k-')
    
    plt.xlim(lims)
    plt.ylim(lims)
    plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))
    plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
    correlation = avana_combined['lfcRAS'].corr(avana_combined['lfcAvana'])
    sns.scatterplot(data=avana_combined[avana_combined['Essentiality'] == 'Others'],
                    x='lfcAvana', y='lfcRAS', color='grey', alpha=0.6, s=24, label='Others')

    sns.scatterplot(data=avana_combined[avana_combined['Essentiality'] == 'Pan-essential genes'],
                    x='lfcAvana', y='lfcRAS', color='red', alpha=0.8, s=24, label='Pan-essential genes')
    
    sns.scatterplot(data=avana_combined[avana_combined['Essentiality'] == 'Non-essential genes'],
                    x='lfcAvana', y='lfcRAS', color='blue', alpha=0.8, s=24, label='Non-essential genes')

    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3)
    plt.xlabel('Avana mean LFC in corresponding cell line')
    plt.ylabel('Mean LFC in our screen')
    plt.title(f'Comparison to AVANA for {cur_sample} (Cor = {round(correlation,2)})',size=15)
    
    plt.gca().spines['top'].set_visible(True)
    plt.gca().spines['right'].set_visible(True)
    plt.gca().spines['bottom'].set_visible(True)
    plt.gca().spines['left'].set_visible(True)
    plt.gca().spines['top'].set_color('black')
    plt.gca().spines['right'].set_color('black')
    plt.gca().spines['bottom'].set_color('black')
    plt.gca().spines['left'].set_color('black')
   
    
    plt.gca().set_facecolor('white')
    #plt.gca().patch.set_alpha(0)

    plt.grid(False)
    
    
    plt.savefig(f"QualityControl/UsePlatepDNA/CompareToAVANA/CompareToAVANA_{cur_sample}.pdf")
    plt.close()
```

```python

def transform_dataSeq(filtered_controls, type_label):
    filtered_data = LFCSequencesPerTarget.loc[
    LFCSequencesPerTarget.index.intersection(filtered_controls['GuideTargetSymbol'])]

    filtered_data = filtered_data.reset_index().rename(columns={'index': 'GuideTargetSymbol'})
    melted_data = filtered_data.melt(id_vars=['GuideTargetSymbol'], var_name='Sequence', value_name='logFC')
    melted_data['type'] = type_label
    return melted_data

RasCombinatoryCommonEssentialControls = tc.get(name='ras-library-files-35fa', file='RasCombinatoryCommonEssentialControls')
RasCombinatoryNonEssentialControls = tc.get(name='ras-library-files-35fa', file='RasCombinatoryNonEssentialControls')

common_essential_df = transform_dataSeq(RasCombinatoryCommonEssentialControls, 'CommonEssential')
nonessential_df = transform_dataSeq(RasCombinatoryNonEssentialControls, 'Nonessential')

df_plot = pd.concat([common_essential_df, nonessential_df], ignore_index=True)

print(df_plot)
```

```python
output_dir = 'QualityControl/UsePlatepDNA/EssentialSeparation/PerSequence/'
os.makedirs(output_dir, exist_ok=True)
custom_colors = {
    'CommonEssential': 'red',
    'Nonessential': 'blue'
}
# Iterate through unique sequences and create density plots
for cur_sample in df_plot['Sequence'].unique():
    subset = df_plot[df_plot['Sequence'] == cur_sample]
    
    plt.figure(figsize=(8, 8))
    sns.kdeplot(data=subset, x='logFC', hue='type', fill=True, common_norm=False, alpha=0.4,palette=custom_colors)
    plt.axvline(x=-1, linestyle='--', color='grey')
    
    plt.title(f'{cur_sample}',size=15)
    plt.xlabel('logFC',size=15)
    plt.ylabel('Density',size=15)
    plt.xticks(rotation=90,size=15)
    plt.yticks(size=15)
    plt.legend(title=None, loc='upper center', bbox_to_anchor=(0.5, 1.15), labels=['Nonessential','CommonEssential'],fontsize=15,ncol=2)
    plt.grid(False)
    plt.savefig(os.path.join(output_dir, f'EssentialSeparation_{cur_sample}.pdf'))
    plt.close()
```
