# Agreement model

```python
avana_lfc = tc.get(name='internal-24q4-8c04', file='AvanaLogfoldChange') 
avana_guide_map = tc.get(name='internal-24q4-8c04', file='AvanaGuideMap')
screen_seq_map =  tc.get(name='internal-24q4-8c04',file='ScreenSequenceMap').set_index('SequenceID')
avana_replicate_map = screen_seq_map.loc[avana_lfc.columns]
# collapse replicates to cell lines by median
avana_lfc_cell = avana_lfc.groupby(avana_replicate_map.ModelID, axis=1).median()
```

```python
#copied from DepMap portal code
from scipy.stats import spearmanr, pearsonr
from sklearn.ensemble import RandomForestClassifier
def get_metrics(A, B):
    '''
    computes a bunch of agreement measures between vectors A and B, suitable for seeing
    whether two measurements of gene effect are in agreement
    '''
    assert all(A.index == B.index), "misaligned indices between the two series"
    mask = A.notnull() & B.notnull()
    a = A[mask]
    b = B[mask]
    a = a.loc[b.index]
    meanA = a.mean()
    meanB = b.mean()
    mean_se = (meanA - meanB)**2
    mean_mean = .5*(meanA + meanB)
    corr = a.corr(b)
    spearman = spearmanr(a, b)[0]
    varA = a.var()
    varB = b.var()
    var_se = np.abs(varA - varB)
    var_mean = .5*(varA + varB)
    centered_SE = ((a - meanA - b + meanB) **2).mean()
    status_agreement = ((a < -.5) == (b < -.5)).mean()
    return {
        'mean_se': mean_se,
        'mean_mean': mean_mean,
        'corr': corr,
        'spearman': spearman,
        'var_se': var_se,
        'var_mean': var_mean,
        'centered_SE': centered_SE,
        'max_mean': max(meanA, meanB),
        'min_mean': min(meanA, meanB),
        'max_var': max(varA, varB),
        'min_var': min(varA, varB),
        'max_max': max(max(a), max(b)),
        'min_min': min(min(a), min(b)),
        'status_agreement': status_agreement
    }
```

```python
# generate the agreement metrics for 5000 random pairs of guides targeting the same gene
features = []
is_same = pd.Series(dtype='int64')
nsamples = 5000
genes = np.random.choice(avana_guide_map.Gene.dropna().unique(), size=nsamples, replace=False)
for i, gene in enumerate(genes):
    if not i%1000:
        print('%1.0f%% done' % (i * 100./len(genes)))
    guides = avana_guide_map.query('Gene == "%s"' % gene).sgRNA.tolist()
    if len(guides) < 2:
        continue     
    np.random.shuffle(guides)
    
    out = get_metrics(avana_lfc_cell.loc[guides[0]], avana_lfc_cell.loc[guides[1]])
    out.update({'gene': gene, 'guide1': guides[0], 'guide2': guides[1]})
    features.append(out)
    is_same[gene] = 1
    
# then generate the agreement metrics for 5000 random pairs of guides targeting different genes 
# (using same individual genes as above)
for i, gene in enumerate(genes):
    if not i%1000:
        print('%1.0f%% done' % (i * 100./len(genes)))
    guides = avana_guide_map.query('Gene == "%s"' % gene).sgRNA.tolist()
    if len(guides) < 2:
        continue
        
    gene2 = gene
    while gene2 == gene or (gene + ' x ' + gene2) in is_same:
        gene2 = np.random.choice(genes)
        guide2 = np.random.choice(avana_guide_map.query('Gene == %r' % gene2).sgRNA)

    guide1 = guides[0]
    # avoid the (unlikely) case that the two genes happen to share a guide and 
    # we've chosen that guide for both genes
    while guide1 == guide2:
        guide1 = np.random.choice(guides)
    np.random.shuffle(guides)
    # a human readable name for the pair of chosen genes
    pseudogene = gene + ' x ' + gene2
    
    out = get_metrics(avana_lfc_cell.loc[guide1], avana_lfc_cell.loc[guide2])
    out.update({'gene': pseudogene, 'guide1': guide1, 'guide2': guide2})
    features.append(out)
    is_same[pseudogene] = 0
    
features = pd.DataFrame(features).set_index('gene')

model = RandomForestClassifier(20, min_impurity_decrease=.00001, min_samples_leaf=100, max_depth=6)
model.fit(features.drop(['guide1','guide2'], axis=1).values, is_same.values)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
# Compute in sample separation (training data)
in_sample = model.predict_proba(
    features.drop(['guide1', 'guide2'], axis=1).values, 
)[:, 1]
f, ax = plt.subplots(1, 1, figsize=(3.5,2.5))
sns.kdeplot(in_sample[is_same==1], label="Same Gene", ax=ax, bw_adjust=0.5)
sns.kdeplot(in_sample[is_same==0], label='Different Genes', ax=ax, bw_adjust=0.5)
ax.set_xlabel("Probability guides target the same gene"); ax.legend()
plt.tight_layout();
```

```python
agreement = []
features = []
for i, gene in enumerate(gene_indexed.index.unique()):
    if not i % 1000 and i > 0:
        print('%i genes complete' % i)
    guides = gene_indexed.loc[[gene]]
    if len(guides) < 2:
        continue
    for j, guide1 in enumerate(guides[:-1]):
        for guide2 in guides[j+1:]:
            assert guide1 != guide2, "%r %s %s" % (guides, guide1, guide2)
            features.append(get_metrics(
                full_lfc_seq_sgrna.loc[guide1], 
                full_lfc_seq_sgrna.loc[guide2]
            ))
            agreement.append({'gene': gene, 'guide1': guide1, 'guide2': guide2})
```

```python
# Collect results
features = pd.DataFrame(features)
agreement = pd.DataFrame(agreement)
agreement['agreement'] = model.predict_proba(features.values, )[:, 1]
```

```python
agreement_per_guide = pd.melt(agreement, id_vars=['gene', 'agreement'], value_vars=['guide1', 'guide2'], var_name='guide_type', value_name='guide')
agreement_per_guide = agreement_per_guide.groupby(['guide']).mean('agreement').sort_values(['agreement'])
```
